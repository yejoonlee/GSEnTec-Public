```python
import pandas as pd
```

```python
df_topic = pd.read_csv(r'./연도_내용_토픽.csv', encoding='utf-8')
df_topic.drop(['Unnamed: 0'], axis='columns', inplace=True)    
display(df_topic)
```

```python
topics_2015 = df_topic.loc[df_topic['년도'] == 2015]
topics_2016 = df_topic.loc[df_topic['년도'] == 2016]
topics_2017 = df_topic.loc[df_topic['년도'] == 2017]
topics_2018 = df_topic.loc[df_topic['년도'] == 2018]
topics_2019 = df_topic.loc[df_topic['년도'] == 2019]

topics_2018 = topics_2018.reset_index()
display(topics_2018['가장 비중이 높은 토픽'])
```

```python
df_surv_sheet1 = pd.read_excel(r'./data/조직역량서베이/2018년/2018_GS엔텍_설문조사_결과_수정.xlsx', encoding='utf-8')
# df_surv.drop(['Unnamed: 0'], axis='columns', inplace=True)    
display(df_surv_sheet1)
```

```python
q_t = pd.concat([df_surv_sheet1, topics_2018['가장 비중이 높은 토픽']], axis = 1)
display(q_t)
```

```python
df_surv_sheet2_p = pd.read_excel(r'./data/조직역량서베이/2018년/2018_GS엔텍_설문조사_결과_수정.xlsx',sheet_name = 'GS엔텍', encoding='utf-8')
# df_surv.drop(['직원id','essay71'], axis='columns', inplace=True)    
# display(df_surv_sheet2)
df_surv_sheet2 = df_surv_sheet2_p.loc[3:]
display(df_surv_sheet2)
list(df_surv_sheet2_p.loc[2])
# df_surv_sheet2.reindex(columns=list(df_surv_sheet2_p.loc[2]))
df_surv_sheet2.columns = list(df_surv_sheet2_p.loc[2])
display(df_surv_sheet2)

# df.reindex(columns=['http_status', 'user_agent'])
```

```python
dict = {}
for i, code_re in enumerate(q_t['code']):
#     print(i, code_re)
    topic = q_t.loc[i, '가장 비중이 높은 토픽'] 
    
    if code_re not in df_surv_sheet2.columns:
        try:
            scores = list(df_surv_sheet2[code_re+'_R'])
        except:
            print(code_re)
    else:
        scores = list(df_surv_sheet2[code_re])
    dict[code_re] = [topic, scores[:-1]]
print(dict)
```
