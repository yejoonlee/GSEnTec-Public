### 경력 max, 조화평균

```python
from calendar import monthrange
from datetime import datetime, timedelta

import statistics

def readExcel(xlse_path, sheetName):
    xls_file = pd.ExcelFile(xlse_path)
    data = xls_file.parse(sheetName)
    return data

data = readExcel('./data/3. 경력현황_수정.xls','Sheet1')

new_df = data[['직원id','회사명','입사일','퇴사일']]
str_date = data['입사일'].fillna('00/00/00')
end_date = data['퇴사일'].fillna('00/00/00')


def monthdelta(d1, d2):         #개월수 계산 함수
    delta = 0
    while True:
        mdays = monthrange(d1.year, d1.month)[1]
        d1 += timedelta(days=mdays)
        if d1 <= d2:
            delta += 1
        else:
            break
    return delta

dic_month = {}
id = data['직원id']

for i in range(len(data)):

    date1 = str_date.loc[i]
    date2 = end_date.loc[i]

    if date1 == '00/00/00' or date2 == '00/00/00' :
        pass

    else :
        change_d1 = datetime.strptime(str(date1), '%Y/%m/%d')
        change_d2 = datetime.strptime(str(date2), '%Y/%m/%d')
        result = monthdelta(change_d1, change_d2)
        if result == 0 :
            result +=1
        try:
            dic_month[id.loc[i]].append(result)
        except:
            dic_month[id.loc[i]] = [result]

# print(dic_month)

dic_max = {}            # 최대값
for key in dic_month.keys():

    max_month = max(dic_month[key])
    dic_max[key] = [max_month]

# print(dic_max)

dic_hm = {}             #조화평균
for key in dic_month.keys():
    month = dic_month[key]
    result = statistics.harmonic_mean(month)
    dic_hm[key] = [result]

# print(dic_hm)

for key in dic_max.keys():
    update_db(connection, cursor, 'BASE', 'pp_경력_최대', dic_max[key][0], key)
    update_db(connection, cursor, 'BASE', 'pp_경력_조화', dic_hm[key][0], key)
```
